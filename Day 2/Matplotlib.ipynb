```python
#import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate 100 evenly spaced points between 0 and Ï€/4 . Store them in a 1-D array x.
```

```python
from math import pi
print(pi)
```

```python
x = np.linspace(0, pi/4, 100)
```

```python
# Compute 1-D arrays:
        # y1 = Sin(x)
        # y2 = Cos(x)
        # y3 = Tan(x)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

```python
# Plot the above functions on separate plots.
```

```python
plt.plot(x, 
         y1)

plt.show()
```

```python
plt.plot(x, 
         y2)

plt.show()
```

```python
plt.plot(x, 
         y3)

plt.show()
```

```python
# In each of the plots:
        # Add an appropriate x-label.
        # Add an appropriate y-label.
        # Add an appropriate label to the plot.
        # Display a grid.
        # Set x limit from 0 to Ï€/4 .
        # Set y limit from 0 to 1.1.
```

```python
plt.plot(x, 
         y1)

plt.xlabel('Values between 0 and pi/4')
plt.ylabel('Sin() values')
plt.title('Graph of Sin(x)')
plt.grid(True)
plt.xlim([0, pi/4])
plt.ylim([0, 1.1])

plt.show()
```

```python
plt.plot(x, 
         y2)

plt.xlabel('Values between 0 and pi/4')
plt.ylabel('Cos() values')
plt.title('Graph of Cos(x)')
plt.grid(True)
plt.xlim([0, pi/4])
plt.ylim([0, 1.1])

plt.show()
```

```python
plt.plot(x, 
         y3)

plt.xlabel('Values between 0 and pi/4')
plt.ylabel('Tan() values')
plt.title('Graph of Tan(x)')
plt.grid(True)
plt.xlim([0, pi/4])
plt.ylim([0, 1.1])

plt.show()
```

```python
# Plot all the three functions on the same plot using:
        # Red color and a solid line for Sin(x).
        # Blue color and a dotted line for Cos(x).
        # Green color and a dashed line for Tan(x).
        # Add an appropriate label, x-label, y-label and set limits for the graph.
```

```python
plt.plot(x, y1, color='r', 
         label='Sin()')
plt.plot(x, y2, linestyle=':', 
         color='b', label='Cos()')
plt.plot(x, y3, linestyle='--', 
         color='g', label='Tan()')

plt.xlabel('Values between 0 and pi/4')
plt.ylabel('Trignometric values')
plt.title('Sin(x), Cos(x) and Tan(x)')
plt.grid(True)
plt.xlim([0, pi/4])
plt.ylim([0, 1.1])

plt.legend()

plt.show()
```

```python
# Matplotlib(Scatter and Bar plots)
```

```python
# Consider the annual revenues of the four big companies(Amazon, Meta, Microsoft and Apple) for the years 2018-2021.
# Create several lists that include the names of these companies with yearly revenues(in billion dollars) such as:
        # company = ['Amazon', 'Meta', 'Microsoft', 'Apple']
        # revenue_2018 = [233, 56, 110, 265]
        # revenue_2019 = [280, 71, 123, 260]
        # revenue_2020 = [386, 86, 143, 274]
        # revenue_2021 = [471, 118, 168, 366]
```

```python
company = ['Amazon', 'Meta', 'Microsoft', 'Apple']
revenue_2018 = [233, 56, 110, 265]
revenue_2019 = [280, 71, 123, 260]
revenue_2020 = [386, 86, 143, 274]
revenue_2021 = [471, 118, 168, 366]
```

```python
# Draw a scatter plot using the â€˜scatterâ€™ function in matplotlib and color the revenue points.
```

```python
plt.scatter(company, revenue_2018, 
            color='hotpink')

plt.show()
```

```python
plt.scatter(company, revenue_2019, 
            color='#88c999')

plt.show()
```

```python
colors = np.array(["red","green","blue","yellow"])
plt.scatter(company, revenue_2019, 
            c=colors)

plt.show()
```

```python
# Draw a bar plot for each year separately.
```

```python
plt.bar(company, revenue_2018, color ='blue',
        width = 0.1)
 
plt.xlabel("Company name")
plt.ylabel("Revenue generated in 2018")
plt.title("Revenue generated by companies in 2018")

plt.show()
```

```python
plt.bar(company, revenue_2019, color ='maroon',
        width = 0.2)
 
plt.xlabel("Company name")
plt.ylabel("Revenue generated in 2019")
plt.title("Revenue generated by companies in 2019")

plt.show()
```

```python
plt.bar(company, revenue_2020, color ='green',
        width = 0.3)
 
plt.xlabel("Company name")
plt.ylabel("Revenue generated in 2020")
plt.title("Revenue generated by companies in 2020")

plt.show()
```

```python
plt.bar(company, revenue_2021, color ='orange',
        width = 0.4)
 
plt.xlabel("Company name")
plt.ylabel("Revenue generated in 2021")
plt.title("Revenue generated by companies in 2021")

plt.show()
```

```python
# Draw four sub-plots corresponding to revenue for each year in one figure.
```

```python
plt.subplot(2,2,1)
plt.bar(company, revenue_2018, color ='blue',
        width = 0.4)

plt.subplot(2,2,2)
plt.bar(company, revenue_2019, color ='maroon',
        width = 0.4)

plt.subplot(2,2,3)
plt.bar(company, revenue_2020, color ='green',
        width = 0.4)

plt.subplot(2,2,4)
plt.bar(company, revenue_2021, color ='orange',
        width = 0.4)
```

```python
# Create a stacked bar chart for each company.
```

```python
plt.bar(company, revenue_2018, color='r')
plt.bar(company, revenue_2019, bottom=revenue_2018, color='b')
plt.bar(company, revenue_2020, bottom=revenue_2019, color='g')
plt.bar(company, revenue_2021, bottom=revenue_2020, color='y')

plt.show()
```
